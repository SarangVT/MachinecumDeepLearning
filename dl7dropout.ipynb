```python
"""In deep learning, the vanishing gradient problem, is encountered:
1. In deep NN, having multiple hidden layers.
2. Where activation function is sigmoid/tanh.
As, 0.1**4=0.0001, so the product of large number of derivatives, vanish out.

How to recognize ?
1. Loss is having no significant change.
2. Weights are not having significant change.

How to handle vanishing gradient?
1. Reduce Model Complexity: Minimize number of hidden layers, make the network shallow.
2. Using ReLU Activation Functions: max(0,Z), squeezes values in negative direction.
Derivatives: Either 0 or 1, so muktiplying several 1 results in 1 only.
Dying ReLU Problem: It is because sometimes it becomes 0 as by multiplying by zero, hence again updation terminates, so we use leaky ReLU.
3. Proper weight intialization
4. Batch Normalization
5. Residual Network


Exploding Gradient Problem: Sometimes gradients are increased by multiplying, several gradient greater than 1, it leads to random behaviour, found especially in RNNs.
"""
```

```python
"""Techniques to Improve Neural Network:
1. Fune tuning NN hyperparameters:
E.g. Number of hidden layers, number of neurons, learning rate, optimizer, in MBGD batch size, activation function, numnber of epochs.
2. By solving problems:
	Vanishing / Exploding gradient
    Not enough data
    Slow training
    Overfitting

Number of hidden layers: Keep a balance, but focus on more number of hidden layers with less neuron, as deep learning is based on representative learning, and to enable transfer learning, but stop adding layers once reached overfitting.
Number of Neurons: Sufficient, generally starting with a bit more is fine.
Batch Size: Smaller(8-32), generalization, results better, but slow.
Larger(8192), are fast, but sometimes not stable, but we can use learning rate scheduler that increases performance, generally increasing.
Based on GPU RAM.
Epochs: When implemented with early stopping, but keep as many as possible, according to available time.

Vanishing / Exploding Gradients: Weight intialization, activation function, batch normalization, gradient clipping.
Not enough data: Transfer learning, unsupervised learning.
Slow training: Optimizers, learning-rate scheduler.
Overfitting: L1 and L2 regularization, Dropouts.
"""
```

```python
"""Early Stopping:
tensorflow.keras.callbacks.EarlyStopping{
	monitor = "val_loss" //parameter to be monitored
    min_delta=0 //minimum quantity below which it is considered no improvement
	patience=3 //number of epochs until which training stops
    verbose=1 //1: tells that it stopped due to early stopping, 1 : tells about it
    mode = auto //min: when quantity has stopped decreasing; max: when quantity has stopped increasing; auto:identifies direction according to monitored quantity
    baseline: training will stop if model doesn't show improvement over the monitored quantity
    restore_best_weights: false //false: weights at current epoch is considered ; true: weights of the epoch with best performance are restored, regardless of baseline, if no epoch improves on baseline, training will run for patience epochs and restore weights from the best in the set.    
}

Normalizing Inputs: Required for inputs, with diverse range of scale, like salary ranging from 0 to 10000000000
Simple Explanation: It focuses more on the changes in gradient descent of large values, thus ignoring smaller features.
Geometric Intuition: When we plot a 3-D contour or hyperplane of the all parameters, the normalized data distribution has symmetric advantage leading to better convergence instead of unnormalized distribution having unsymmetric geometry.
Standardization of data,
xi = (xi - µ)/σ, where σ is standard deviation and µ is mean.
It means first standard deviation is made 1(doesn't depend on -µ) and then mean is made 0, by subtracting mean n times, n is total entries in data.
Generally, by this data comes from -1 to 1.

Normalization of data:
xi = (xi - xmin) / (xmax - xmin)
It basically encloses it in an interval of [0,1].

If we know, the maximum and minimum values we would prefer normalization but if we don't know standardize it.
Standardization is also useful for outliers.
"""
```

```python
"""Problem of Overfitting:
1. Add more Data
2. Reduce Complexity
3. Early stopping
4. Regularization: L1 and L2
5. Dropout"""
```

```python
"""Dropouts: 
Here, we drop nodes in hidden layers, randomly during each epoch, so everytime a new neural network is achieved.
They are very much like random forest, they form multiple decision trees and take deceision based on majority.
And so, here there are multiple NN, and ultimately all weights are considered and taken according to their probability W(1-P), where P is their dropout ratio, that is the probability they are switched off.
Overfitting: Dropout ratio increase
First try on last layer
CNN: 0.4-0.5, RNN:0.2-0.3, ANN: 0.1-0.5
Convergence has delay due to dropout.
As, due to change in loss function, the gradient calculation changes.
"""
```

```python
# Generate Data
X_train = np.linspace(-1, 1, 20)
y_train = np.array([-0.6561 , -0.3099 , -0.59035, -0.50855, -0.285  , 
                    -0.2443 , -0.02445,  0.00135, -0.2006 ,  0.07475, 
                    -0.1422 ,  0.06515,  0.15265,  0.3521 ,  0.28415,  
                    0.5524 ,  0.23115,  0.20835, 0.4211,  0.60485])
X_test = np.linspace(-1, 1, 20)
y_test = np.array([-0.69415, -0.451  , -0.43005, -0.4484 , -0.1475 ,
                   -0.5019 , -0.28055,  0.24595, -0.21425, -0.0286 ,  
                   0.23415,  0.46575, 0.07955,  0.1973 ,  0.0719 ,
                   0.3639 ,  0.5536 ,  0.3365 , 0.50705,  0.33435])
```
